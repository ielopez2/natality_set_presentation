```sas
*data wrangeling: creating variable that counts complication occurances at labor, and the variable indicating risk factor present (med_cond);
*Getting rid of unknown values as well.;
data sectA;
 set cmp01 cmp02 cmp03 cmp04 cmp05 cmp06 cmp07 cmp08 cmp09 cmp10 cmp11 cmp12 cmp13 cmp14 cmp15 cmp16 cmp17 cmp18 cmp19 cmp20;
 if (MM_AICU = 'Y' or 
     MM_MTR = 'Y' or 
     MM_RUPT = 'Y' or 
     MM_PLAC = 'Y' or 
     MM_UHYST = 'Y' or 
     ME_PRES = '2' or 
     ME_PRES = '3') 
 then comp_occur = 1;
  else comp_occur = 0;
  if (RF_PDIAB = 'Y' or
      RF_GDIAB = 'Y' or
      RF_PHYPE = 'Y' or
      RF_GHYPE = 'Y' or
      RF_EHYPE = 'Y' or
      BMI_R = '1' or
      BMI_R = '6')
    then med_cond = 1;
    else med_cond = 0;
 where MM_AICU ne 'U' and 
     MM_MTR ne 'U' and 
     MM_RUPT ne 'U' and 
     MM_PLAC ne 'U' and 
     MM_UHYST ne 'U' and 
     ME_PRES ne 9 and 
     WTGAIN ne 99 and 
     RF_PDIAB ne 'U' and 
     RF_GDIAB ne 'U' and 
     RF_PHYPE ne 'U' and 
     RF_GHYPE ne 'U' and 
     RF_EHYPE ne 'U' and
     BMI_R ne 9 and
     PRECARE ne 99;
run;


proc print data = sectA (obs = 10);
run;


*assiging bins to variables;
data sectA;
    set sectA;
 if RF_PDIAB = 'Y' then RF_PDIAB_bin=1;
  else RF_PDIAB_bin =0;
 if RF_GDIAB="Y" then RF_GDIAB_bin=1;
  else RF_GDIAB_bin=0;
 if RF_PHYPE='Y' then RF_PHYPE_bin=1;
  else RF_PHYPE_bin=0;
 if RF_GHYPE="Y" then RF_GHYPE_bin=1;
  else RF_GHYPE_bin=0;
 if RF_EHYPE = "Y" then RF_EHYPE_bin = 1;
  else RF_EHYPE_bin = 0;
 if MM_AICU = "Y" then MM_AICU_bin = 1;
  else MM_AICU_bin = 0;
 if MM_MTR = "Y" then MM_MTR_bin = 1;
  else MM_MTR_bin = 0;
 if MM_RUPT = "Y" then MM_RUPT_bin = 1;
  else MM_RUPT_bin = 0;
 if MM_PLAC = "Y" then MM_PLAC_bin = 1;
  else MM_PLAC_bin = 0;
 if MM_UHYST = "Y" then MM_UHYST_bin = 1;
  else MM_UHYST_bin = 0;
 if PRECARE = '00' then precare_recvd = 0;
  else precare_recvd = 1;
run;

*cases where mothers had complication during labor.;
data sectA_comp;
    set sectA;
    where comp_occur = 1;
run;

*This dataset will only include mothers with risk factors present;
data sectA_med;
    set sectA;
    where med_cond = 1;
run;
```

```sas
proc freq data = sectA;
    table comp_occur*med_cond;
run;
```

```sas
*logistic regression used in an attempt to predicted complications occuring during labor.;
*use ctable to create classification table.;
proc logistic data=sectA ;
 class RF_PDIAB_bin RF_GDIAB_bin RF_PHYPE_bin RF_GHYPE_bin RF_EHYPE_bin BMI_R ME_PRES WTGAIN / param=ref ref=first;
 model comp_occur(event="1") = RF_PDIAB_bin RF_GDIAB_bin RF_PHYPE_bin RF_GHYPE_bin RF_EHYPE_bin / rsquare;
 *ctable;
run;

*correlation plot used to check for collinearity using spearman correlation.;
proc corr data = sectA spearman;
 var RF_PDIAB_bin RF_GDIAB_bin RF_PHYPE_bin RF_GHYPE_bin RF_EHYPE_bin comp_occur;
run;

proc freq data = sectA_med;
    table comp_occur*med_cond;
run;

proc sgplot data=sectA_med;
 vbar precare_recvd / group=med_cond groupdisplay=cluster;
run;

proc freq data = sectA_med;
    table precare_recvd;
run;

proc sgplot data=sectA_med;
 vbar comp_occur / group = precare_recvd groupdisplay=cluster;
run;

*use spearman correlation, >.90 will indicate collinearity. May happen because people with risk factors ussually have others.;
*c is the area under the curve, which is indicating my response is being obtaining randomly.;
*look at confusion- classification table to check for missclasification;
*create bar plot with population cities and which ones are lacking in mothers who go to prenatal care.
```

```sas
*bar plot used in presentation, shows that there should be more mothers starting prenatal care earlier.;
proc sgplot data=sectA_med NOWALL noborder;
 hbar precare / datalabel  fillattrs=(color=blue) datalabelattrs=(size=12pt color=blue) ;
 inset (" By first 2 months, only 48.69% of risk factored mothers are accounted for"="") / textattrs = (SIZE=10 COLOR='blue');
run;

proc freq data = sectA_med;
    table precare / plots =freqplot ;
run;
```
